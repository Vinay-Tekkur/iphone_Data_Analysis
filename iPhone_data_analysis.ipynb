```python
import pandas as pd
```

```python
df_iphonedata = pd.read_csv('iphone.csv')
```

```python
df_iphonedata.head()
```

```python
# 1 - The column names have spaces . rename the column names to have underscore '_' 
# instead of space (try to do in one go instead of specifying each column nam in rename method)
df_iphonedata.columns = ['product_name', 'product_url', 'brand', 'sale_price', 'mrp',
       'number_of_ratings', 'number_of_reviews', 'upc', 'star_rating', 'ram']


```

```python
df_iphonedata.head()
```

```python
# 2- start rating for some of the models is missing in the dataset. 
# fill those missing values with the average rating all the models.
avg_product_rating = round(df_iphonedata['star_rating'].mean(),2)
df_iphonedata['star_rating'].fillna(avg_product_rating)
```

```python
df_iphonedata['star_rating'].isnull()
```

```python
# 3- Now instead of filling missing values with avg rating of full dataset ,
# fill with avg rating based on RAM. example : 
# if rating for a 2 gb phone is missing then take average of all other 2 gb phones rating and fill that value. 
list(df_iphonedata['ram'].unique())
df_iphonedata[df_iphonedata['ram'] == '2 GB']  

fc = df_iphonedata['ram'] == '2 GB'
avg_rating = round(df_iphonedata.loc[fc,'star_rating'].mean(),1)
avg_rating
df_iphonedata.loc[fc,'star_rating'].fillna(avg_rating, inplace=True)
```

```python
def get_avg_ratings(total_rams):
    for ram in total_rams:
        fc = df_iphonedata['ram'] == ram
        avg_rating = round(df_iphonedata.loc[fc,'star_rating'].mean(),1)
        df_iphonedata.loc[fc,'star_rating'].fillna(avg_rating, inplace=True)
    
ram_lst = list(df_iphonedata['ram'].unique())
get_avg_ratings(ram_lst)
```

```python
# 4- create a new column in the dataframe "Discount_Percentage" based on MRP and sale value
df_iphonedata['discount_percentage'] = round((df_iphonedata['mrp'] - df_iphonedata['sale_price']) / df_iphonedata['mrp'],2) * 100
```

```python
df_iphonedata.head()
```

```python
# 5- which model has highest percent discount ?
high_discount_prd = df_iphonedata.sort_values('discount_percentage',ascending=False).head(1)
high_discount_prd
```

```python
# 14- which iphone with 128 GB space has highest ratings to review ratio
get_iphone_storage = df_iphonedata['product_name'].str.split(pat=',',n=1,expand=True)
get_iphone_storage[1] = get_iphone_storage[1].str.replace(')','')
get_iphone_storage[1] = get_iphone_storage[1].str.strip()
df_iphonedata['storage'] = get_iphone_storage[1]
df_iphonedata.head()
```

```python
iphone_128GB = df_iphonedata[df_iphonedata['storage'] == '128 GB']
rating_reviw_ratio = iphone_128GB.groupby('product_name').agg({'number_of_ratings':'max','number_of_reviews':'max'})
rating_reviw_ratio['ratio'] = rating_reviw_ratio['number_of_ratings'] / rating_reviw_ratio['number_of_reviews']
```

```python
rating_reviw_ratio.sort_values('ratio',ascending=False,inplace=True)
rating_reviw_ratio.head(1)
```
